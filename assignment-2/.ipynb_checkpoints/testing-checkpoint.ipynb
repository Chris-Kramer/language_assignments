```python
#Vi laver en mere kompleks tokenizer. 
#A quick regex tokenizer for splitting files.
def tokenize(input_string):
    #Split on any non-alphanumeric character.
    #tokenizer = re.compile(r"[\W+]")#r først viser, at det er en regex
    #Tokenize
    input_string = input_string.lower()
    token_list = re.findall(r'\b\w+\b', input_string)
    #token_list = tokenizer.split(input_string)
    return token_list
```

```python
def kwic(text, keyword, window_size):
    lines = ""
    for match in re.finditer(keyword, text):
        #First character of match
        word_start = match.start()
        #last character index of match
        word_end = match.end()
        
        #left window
        left_window_start = max(0, word_start - window_size)#If it is negative make it zero.
        left_window = text[left_window_start:word_start]
        
        #RIght window
        right_window_end = word_end + window_size
        right_window = text[word_end : right_window_end]
        
        #print line
        line = f"{left_window} {keyword} {right_window}"
        lines = lines + line
        #lines.append(line)
        
    return lines
```

```python
def find_collocates (text, keyword, window_size):
    
    #Create concordance
    concordances = kwic(text, keyword, window_size)
    collocates = []
    
    #Tokenize each line
    for concordance in concordances:
        tokenized_concordance = tokenize(concordance)
        
        #add token to list of collocates
        for token in tokenized_concordance:
            #if token is the keyword then remove it. 
            if token == keyword:
                tokenized_concordance.remove(token)
            
            else:
                collocates.append(token)
            
    return collocates
```

```python
collocates = []
#Create a list of collocates
for token in tokenized_lines:
    if token not in collocates:
        collocates.append(token)

collocate_counts = []
for collocate in collocates:
    count = tokenized_lines.count(collocate)
    collocate_counts.append(count)
    #print(f"Word: {collocate}, count: {count}")
```

```python
u = tokenized_corpus.count(keyword) #how often does the keyword appear?
N = len(tokenized_corpus) #Lenght of text

index = 0 #index
for collocate in collocates:
    print(index)
    v = tokenized_corpus.count(collocate) #How often does this collocate appear?
    
    O11 = collocate_counts[index] #How often does this collocate and keyword appear?
    O12 = u - collocate_counts[index] #How often does the keyword appear without this collocate?
    O21 = v - O11 #How often does this collocate appear without keyword
    
    #If we only found a partial word as a collocate the if-statement ignores it.
    # It basically checks whether or not there are more collocates than tokens.
    # If this is true, that means, that we have a partial word, which will spew out strange results and should, therefore, be ignored.
    if O21 < 0:
        continue
    
    else: 
        R1 = O11 + O12 #Calculate R1
        C1 = O11 + O21 #Calculate C1
    
        E11 = (R1*C1/N) 
    
        #Calculate MI
        MI = math.log(O11/E11)

        print(f"Word: '{collocates[index]}', V: {v} O11: {O11}, O12: {O12}, O21: {O21}, N: {N}")
        print(f"E11: {E11}, MI {MI}")
        index = index + 1
#O12
```
